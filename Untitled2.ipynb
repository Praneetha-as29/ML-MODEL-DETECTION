```python
import shutil
from pathlib import Path
import random

random.seed(42)

# Paths to your images
with_img_folder = Path('dataset/seatbelts/with')
without_img_folder = Path('dataset/seatbelts/without')

# Target directories
for folder in [
    'dataset/seatbelts/images/train',
    'dataset/seatbelts/images/val',
    'dataset/seatbelts/labels/train',
    'dataset/seatbelts/labels/val'
]:
    Path(folder).mkdir(parents=True, exist_ok=True)

def split_and_copy(img_folder, label, train_ratio=0.8):
    images = list(img_folder.glob('*.*'))
    random.shuffle(images)
    split_idx = int(len(images) * train_ratio)
    train_imgs = images[:split_idx]
    val_imgs = images[split_idx:]

    # Create dummy label file for each image (0 or 1 depending on label)
    def create_dummy_label(img_path, label_value, target_label_folder):
        label_file = target_label_folder / f'{img_path.stem}.txt'
        # Dummy box covering whole image: class_id x_center y_center width height (all normalized)
        # You can later replace with real labels
        label_file.write_text(f"{label_value} 0.5 0.5 1.0 1.0\n")

    for img_path in train_imgs:
        shutil.copy(img_path, f'dataset/seatbelts/images/train/{img_path.name}')
        create_dummy_label(img_path, label, Path('dataset/seatbelts/labels/train'))

    for img_path in val_imgs:
        shutil.copy(img_path, f'dataset/seatbelts/images/val/{img_path.name}')
        create_dummy_label(img_path, label, Path('dataset/seatbelts/labels/val'))

# 0 = with seatbelt
split_and_copy(with_img_folder, label=0)
# 1 = without seatbelt
split_and_copy(without_img_folder, label=1)
```

```python
from pathlib import Path
import random
import numpy as np
import matplotlib.pyplot as plt
from ultralytics import YOLO

# ── CONFIG ──────────────────────────────────────────────────────────────
MODEL_PATH   = "runs/detect/train/weights/best.pt"   # <- change if needed
N_PER_CLASS  = 50    # 50 WITH + 50 WITHOUT  = 100 total
SEED         = 42    # reproducible sampling

IMG_DIRS  = [Path("dataset/seatbelts/images/train"),
             Path("dataset/seatbelts/images/val")]
LBL_DIRS  = [Path("dataset/seatbelts/labels/train"),
             Path("dataset/seatbelts/labels/val")]
# ────────────────────────────────────────────────────────────────────────

model = YOLO(MODEL_PATH)
random.seed(SEED)

# ── helpers ─────────────────────────────────────────────────────────────
def get_gt(lbl_path: Path):
    """Return 0 (WITH) / 1 (WITHOUT) or None if label missing/empty."""
    if not lbl_path.exists():
        return None
    txt = lbl_path.read_text().strip().split()
    return int(txt[0]) if txt else None

def top_pred(result):
    """Most-confident predicted class from YOLO result."""
    if len(result.boxes) == 0:
        return None
    idx = result.boxes.conf.argmax()
    return int(result.boxes.cls[idx])

# ── 1. collect paths by class ───────────────────────────────────────────
pos, neg = [], []  # pos = WITH (class 0) , neg = WITHOUT (class 1)

for img_dir, lbl_dir in zip(IMG_DIRS, LBL_DIRS):
    for img_path in img_dir.glob("*.*"):
        lbl_path = lbl_dir / f"{img_path.stem}.txt"
        gt = get_gt(lbl_path)
        if gt == 0:   pos.append((img_path, lbl_path))
        elif gt == 1: neg.append((img_path, lbl_path))

assert len(pos) >= N_PER_CLASS and len(neg) >= N_PER_CLASS, \
       "Not enough images in one of the classes."

sample = random.sample(pos, N_PER_CLASS) + random.sample(neg, N_PER_CLASS)
random.shuffle(sample)

# ── 2. evaluate sample ─────────────────────────────────────────────────
TP = TN = FP = FN = 0

for img_path, lbl_path in sample:
    gt   = get_gt(lbl_path)
    pred = top_pred(model.predict(str(img_path), imgsz=640,
                                  conf=0.25, verbose=False)[0])
    if gt == 0:                      # actual WITH (positive)
        if pred == 0: TP += 1
        else:         FN += 1
    elif gt == 1:                    # actual WITHOUT (negative)
        if pred == 1: TN += 1
        else:         FP += 1

print("===== 100-image balanced sample =====")
print(f"TP: {TP}   TN: {TN}   FP: {FP}   FN: {FN}")

# ── 3. BAR-CHART of counts ─────────────────────────────────────────────
labels  = ["TP", "TN", "FP", "FN"]
counts  = [TP, TN, FP, FN]
colors  = ['green', 'blue', 'red', 'orange']

plt.figure(figsize=(7,4))
bars = plt.bar(labels, counts, color=colors)
plt.title("Confusion-Matrix Counts (100-image sample)")
plt.ylabel("Images")

for bar in bars:
    h = bar.get_height()
    plt.text(bar.get_x()+bar.get_width()/2, h+0.5, h,
             ha='center', va='bottom', fontsize=11)
plt.tight_layout()
plt.show()

# ── 4. HEAT-MAP confusion matrix ───────────────────────────────────────
cm = np.array([[TP, FP],
               [FN, TN]])  # rows=Actual (0,1)  cols=Pred (0,1)

fig, ax = plt.subplots(figsize=(4.5,4.5))
im = ax.imshow(cm, cmap=plt.cm.Blues)
fig.colorbar(im, ax=ax)

ax.set_xticks([0,1]); ax.set_yticks([0,1])
ax.set_xticklabels(["Pred WITH", "Pred WITHOUT"])
ax.set_yticklabels(["Actual WITH", "Actual WITHOUT"])
ax.set_title("Confusion Matrix (heat-map)")

th = cm.max() / 2.0
for i in range(2):
    for j in range(2):
        ax.text(j, i, cm[i,j],
                ha="center", va="center",
                color="white" if cm[i,j] > th else "black",
                fontsize=13, weight='bold')
plt.tight_layout()
plt.show()
```

```python
from pathlib import Path
from ultralytics import YOLO
import matplotlib.pyplot as plt
import numpy as np
import itertools

# Load your trained model
model = YOLO("runs/detect/train/weights/best.pt")  # or your .pt file

# Collect all images from both train and val
image_dirs = [
    Path("dataset/seatbelts/images/train"),
    Path("dataset/seatbelts/images/val")
]

label_dirs = [
    Path("dataset/seatbelts/labels/train"),
    Path("dataset/seatbelts/labels/val")
]

# Helper functions
def get_ground_truth_class(label_path):
    if not label_path.exists():
        return None
    line = label_path.read_text().strip().split()
    return int(line[0]) if line else None

def get_top_prediction(result):
    if len(result.boxes) == 0:
        return None
    i = result.boxes.conf.argmax()
    return int(result.boxes.cls[i])

# Init confusion matrix counters
TP = TN = FP = FN = 0

# Loop over train and val images
for img_dir, label_dir in zip(image_dirs, label_dirs):
    for img_path in img_dir.glob("*.*"):
        label_path = label_dir / f"{img_path.stem}.txt"
        gt = get_ground_truth_class(label_path)
        if gt is None:
            continue

        result = model.predict(source=str(img_path), imgsz=640, conf=0.25, verbose=False)[0]
        pred = get_top_prediction(result)

        if gt == 0:  # Actual: with seatbelt
            if pred == 0:
                TP += 1
            else:
                FN += 1
        elif gt == 1:  # Actual: without seatbelt
            if pred == 1:
                TN += 1
            else:
                FP += 1

# Show raw counts
print(f"True Positive (TP): {TP}")
print(f"True Negative (TN): {TN}")
print(f"False Positive (FP): {FP}")
print(f"False Negative (FN): {FN}")

# Plot confusion matrix
cm = np.array([[TP, FN],
               [FP, TN]])

class_names = ["With", "Without"]

fig, ax = plt.subplots(figsize=(5, 4))
im = ax.imshow(cm, interpolation="nearest", cmap=plt.cm.Blues)
fig.colorbar(im, ax=ax)

ax.set(
    xticks=np.arange(2),
    yticks=np.arange(2),
    xticklabels=[f"Actual: {c}" for c in class_names],
    yticklabels=[f"Pred: {c}" for c in class_names],
    ylabel="", xlabel=""
)

plt.title("Confusion Matrix (Train + Val)")

thresh = cm.max() / 2.0
for i, j in itertools.product(range(2), range(2)):
    ax.text(j, i, cm[i, j],
            ha="center", va="center",
            color="white" if cm[i, j] > thresh else "black",
            fontsize=12)

plt.tight_layout()
plt.show()
```
